Checking the Outliers in data

```python
numerical_columns = food_df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_columns):
    plt.subplot(len(numerical_columns)//3 + 1, 3, i + 1)  # Arrange plots in a grid
    sns.boxplot(y=food_df[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()

plt.show()
```

```python
q1 = food_df['Year_Birth'].quantile(0.25)
q3 = food_df['Year_Birth'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Capping the outliers
food_df['Year_Birth'] = food_df['Year_Birth'].clip(lower_bound, upper_bound)
```

```python
columns_outlier = ['MntWines', 'MntFruits', 'MntMeatProducts','MntFishProducts','MntSweetProducts','MntSweetProducts','MntGoldProds','NumDealsPurchases',
                  'NumWebPurchases','NumCatalogPurchases','NumWebVisitsMonth']

# Loop through each column and apply the outlier capping process
for column in columns_outlier:
    q1 = food_df[column].quantile(0.25)
    q3 = food_df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Capping the outliers for the column
    food_df[column] = food_df[column].clip(lower=lower_bound, upper=upper_bound)

# Display the updated DataFrame
food_df.head()
```

Filling the Null values

```python
education_mean_income = food_df.groupby('Education')['Income'].mean()
```

```python
food_df['Income'] = food_df.apply(
    lambda row: education_mean_income[row['Education']] if pd.isnull(row['Income']) else row['Income'],
    axis=1
)
```

Feature Engineering

```python
food_df['Age'] = 2024 - food_df['Year_Birth']
food_df.info()
```

Summing up the amount spent columns

```python
food_df['TotalAmount_Spent'] = food_df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
```

Summing up the all Purchase Columns

```python
food_df['Total_Purchases'] = food_df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)
```

Find Family Size

```python
food_df['Family_Size'] = food_df['Kidhome'] + food_df['Teenhome']
```
